### 09. 從 nba 資料庫的 players 資料表依據 heightMeters、weightKilograms 以及下列公式衍生計算欄位 bmi，並使用 ROUND 函數將 bmi 的小數點位數調整為 2 位
* 練習使用ROUND(data, digit)，調整取出的資料的位數

```python
calculate_rounded_bmi_from_players =\
"""
-- SQL 查詢語法起點

SELECT heightMeters,
       weightKilograms,
       ROUND(weightKilograms / (heightMeters * heightMeters), 2) AS bmi
  FROM players;

-- SQL 查詢語法終點
"""

rounded_bmi_from_players = pd.read_sql(calculate_rounded_bmi_from_players, conn)
rounded_bmi_from_players
```

### 10. 從 nba 資料庫的 career_summaries 資料表中依據 assists、turnovers 欄位以及下列公式衍生計算助攻失誤比，使用 CAST 函數讓衍生計算欄位的資料類型為浮點數 REAL
* Assists Turnover Ratio = Assists / Turnovers
* 練習使用CAST(data AS given_data_type)將資料轉換型態

```python
calculate_ast_to_ratio_from_career_summaries =\
"""
-- SQL 查詢語法起點

SELECT assists,
       turnovers,
       CAST(assists AS REAL) / turnovers AS assist_turnover_ratio
  FROM career_summaries;
       
-- SQL 查詢語法終點
"""

ast_to_ratio_from_career_summaries = pd.read_sql(calculate_ast_to_ratio_from_career_summaries, conn)
ast_to_ratio_from_career_summaries
```

### 11. 從 covid19 資料庫的 time_series 資料表依據 Date 變數，使用 STRFTIME 函數查詢時間序列資料有哪些不重複的月份

* 練習使用 STRFTIME(Format, data) 處理日期時間資料

```python
find_distinct_year_month_from_time_series =\
"""
-- SQL 查詢語法起點

SELECT DISTINCT STRFTIME('%Y-%m' , Date) AS distinct_year_month
  FROM time_series;

-- SQL 查詢語法終點
"""

distinct_year_month_from_time_series = pd.read_sql(find_distinct_year_month_from_time_series, conn)
distinct_year_month_from_time_series
```

### 12. 從 twElection2020 資料庫的 presidential 資料表利用聚合函數彙總有多少人參與了總統副總統的投票選舉，參考下列的預期查詢結果。

* 練習使用 COUNT(data)來計算欄位總和

```python
summarize_total_votes_from_presidential =\
"""
-- SQL 查詢語法起點

SELECT SUM(votes) AS total_presidential_votes
  FROM presidential;

-- SQL 查詢語法終點
"""

total_votes_from_presidential = pd.read_sql(summarize_total_votes_from_presidential, conn)
total_votes_from_presidential
```

### 13. 從 covid19 資料庫的 daily_report 資料表利用聚合函數彙總截至 2021-03-31 全世界總確診數、總痊癒數以及總死亡數

* 練習使用 SUM(data) 計算欄位總和

```python
summarize_totals_from_daily_report =\
"""
-- SQL 查詢語法起點

SELECT SUM(Confirmed) AS total_confirmed,
       SUM(Recovered) AS total_recovered,
       SUM(Deaths) AS total_deaths
  FROM daily_report;

-- SQL 查詢語法終點
"""

totals_from_daily_report = pd.read_sql(summarize_totals_from_daily_report, conn)
totals_from_daily_report
```
