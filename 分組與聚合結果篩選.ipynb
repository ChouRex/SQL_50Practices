```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('nba.db')
conn.execute("""ATTACH 'covid19.db' AS covid19""")
conn.execute("""ATTACH 'twElection2020.db' AS twElection2020""")
conn.execute("""ATTACH 'imdb.db' AS imdb""")
```

###### Overview
1. 使用 GROUP BY 取代 DISTINCT + ORDER BY，並搭配函式，計算出每個群體的數量
2. 使用 HAVIGN 對取出的資料進行篩選

###### 備註
* 篩選資料有兩種方式 : WHERE, HAVING 
1. WHERE -> 在資料庫撈資料時就啟動的篩選機制
2. HAVING -> 撈出來的資料（或許進行運算後），才啟動的篩選機制

### 25. 從 imdb 資料庫的 movies 資料表計算每一年有幾部在 IMDb.com 獲得高評等的經典電影
* 練習使用 GROUP BY，並搭配函式計算每個群體的數量

```python
count_number_of_movies_by_year_from_movies =\
"""
-- SQL 查詢語法起點

SELECT release_year,
       COUNT(*) AS number_of_movies
  FROM movies
 GROUP BY release_year;

-- SQL 查詢語法終點
"""
number_of_movies_by_year_from_movies = pd.read_sql(count_number_of_movies_by_year_from_movies, conn)
number_of_movies_by_year_from_movies
```

### 26. 從 imdb 資料庫的 movies 資料表計算每一年有幾部在 IMDb.com 獲得高評等的經典電影，只顯示電影數在 5 部以上（>= 5）的年份

* 練習使用 GROUP BY 搭配 HAVING 去做資料篩選

```python
count_number_of_movies_by_year_having_from_movies =\
"""
-- SQL 查詢語法起點

SELECT release_year,
       COUNT(*) AS number_of_movies
  FROM movies
 GROUP BY release_year
HAVING number_of_movies >= 5;

-- SQL 查詢語法終點
"""
number_of_movies_by_year_having_from_movies = pd.read_sql(count_number_of_movies_by_year_having_from_movies, conn)
number_of_movies_by_year_having_from_movies
```

### 27. 從 twElection2020 資料庫的 presidential 資料表暸解台灣 2020 總統副總統的選舉結果

* 練習使用GROUP BY 搭配函式計算目標群體的總數

```python
find_summary_by_candidate_id_from_presidential =\
"""
-- SQL 查詢語法起點

SELECT candidate_id,
       SUM(votes) AS total_votes
  FROM presidential
 GROUP BY candidate_id;

-- SQL 查詢語法終點
"""
summary_by_candidate_id_from_presidential = pd.read_sql(find_summary_by_candidate_id_from_presidential, conn)
summary_by_candidate_id_from_presidential
```

### 28. 從 nba 資料庫的 players 資料表根據 country 暸解截至 2021-03-31，NBA 由哪些國家的球員所組成
* 使用GROUP BY 區分國家球員數量

```python
count_number_of_players_by_country_from_players =\
"""
-- SQL 查詢語法起點

SELECT country,
       COUNT(country) AS number_of_players
  FROM players
 GROUP BY country
 ORDER BY number_of_players DESC;

-- SQL 查詢語法終點
"""
number_of_players_by_country_from_players = pd.read_sql(count_number_of_players_by_country_from_players, conn)
number_of_players_by_country_from_players
```

### 29. 從 nba 資料庫的 players 資料表根據 country 暸解截至 2021-03-31，NBA 由哪些國家的球員所組成，只顯示球員數在 2 位以上（>= 2）並在 9 位以下（<=9）的國家
* 使用 GROUP BY + HAVING 篩選出需要的資料

```python
count_number_of_players_by_country_having_from_players =\
"""
-- SQL 查詢語法起點

SELECT country,
       COUNT(country) AS number_of_players
  FROM players
 GROUP BY country
HAVING number_of_players >= 2 AND number_of_players <= 9
 ORDER BY number_of_players DESC;

-- SQL 查詢語法終點
"""

number_of_players_by_country_having_from_players = pd.read_sql(count_number_of_players_by_country_having_from_players, conn)
number_of_players_by_country_having_from_players
```
