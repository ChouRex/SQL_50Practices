```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('nba.db')
conn.execute("""ATTACH 'covid19.db' AS covid19""")
conn.execute("""ATTACH 'twElection2020.db' AS twElection2020""")
conn.execute("""ATTACH 'imdb.db' AS imdb""")
```

###### Overview
* 練習使用 ORDER BY 透過已知欄位、衍生欄位排序 -> 找出「前n高」或是「前n低」的資料

### 14. 從 nba 資料庫的 career_summaries 資料表中依據 ppg（Points per game，場均得分）找出場均得分最高的 10 名球員

* 練習使用保留字 ORDER BY DESC 搭配 LIMIT 找出「前n高」或是「前n低」的資料

```python
find_top_ten_ppg_from_career_summaries =\
"""
-- SQL 查詢語法起點

SELECT personID,
       ppg
  FROM career_summaries
 ORDER BY ppg DESC
 LIMIT 10;

-- SQL 查詢語法終點
"""

top_ten_ppg_from_career_summaries = pd.read_sql(find_top_ten_ppg_from_career_summaries, conn)
top_ten_ppg_from_career_summaries
```

### 15. 從 covid19 資料庫的 time_series 資料表中依據 Daily_Case 找出前十個單日新增確診數最多的日期
* 練習使用 ORDER BY DESC + LIMIT 

```python
find_top_ten_daily_case_from_time_series =\
"""
-- SQL 查詢語法起點

SELECT Date,
       Country_Region,
       Daily_Cases
  FROM time_series
 ORDER BY Daily_Cases DESC
 LIMIT 10;
 
-- SQL 查詢語法終點
"""

top_ten_daily_case_from_time_series = pd.read_sql(find_top_ten_daily_case_from_time_series, conn)
top_ten_daily_case_from_time_series
```

### 16. 從 nba 資料庫的 career_summaries 資料表中依據 assists、turnovers 欄位以及下列公式衍生計算助攻失誤比，並且讓衍生計算欄位的資料類型為浮點數 REAL，找出助攻失誤比最高的前 10 名球員

* Assists Turnover Ratio = Assists / Turnovers
* 練習使用 ORDER BY + LIMIT

```python
find_top_ten_ast_to_ratio_from_career_summaries =\
"""
-- SQL 查詢語法起點

SELECT personID,
       assists,
       turnovers
  FROM career_summaries
 ORDER BY CAST(assists AS REAL) / turnovers DESC
 LIMIT 10;

-- SQL 查詢語法終點
"""

top_ten_ast_to_ratio_from_career_summaries = pd.read_sql(find_top_ten_ast_to_ratio_from_career_summaries, conn)
top_ten_ast_to_ratio_from_career_summaries
```
