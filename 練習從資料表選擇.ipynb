```python
import sqlite3
import sqlite3
import unittest
import json
import os
import numpy as np
import pandas as pd
```

```python
# 連結資料庫
conn = sqlite3.connect('nba.db')
conn.execute("""ATTACH 'covid19.db' AS covid19""")
conn.execute("""ATTACH 'twElection2020.db' AS twElection2020""")
conn.execute("""ATTACH 'imdb.db' AS imdb""")
```

##### Overview

1. 從資料表取資料 : SELECT data FROM table LIMIT #;
2. 修改取出資料的欄位名稱 SELECT data AS revised_name...
3. 從資料表取出不重複的資料 SELECT DISTINCT data ...

### 01. 從 twElection2020 資料庫的 admin_regions 資料表選擇所有變數，並且只顯示前五列資料
* 練習 SELECT all_data FROM table LIMIT # 的使用

```python
# SQL 注入語法
select_all_from_admin_regions =\
"""
-- SQL 查詢語法起點

SELECT *
  FROM admin_regions
 LIMIT 5;

-- SQL 查詢語法終點
"""

all_from_admin_regions = pd.read_sql(select_all_from_admin_regions, conn)
all_from_admin_regions
```

### 02. 從 nba 資料庫的球隊資料表 teams 中選擇 confName、divName、fullName 三個變數，並且只顯示前十列資料
* 練習 SELECT given_data FROM table LIMIT # 的使用

```python
select_variables_from_teams =\
"""
-- SQL 查詢語法起點

SELECT confName,
       divName,
       fullName
  FROM teams
 LIMIT 10;

-- SQL 查詢語法終點
"""

variables_from_teams = pd.read_sql(select_variables_from_teams, conn)
variables_from_teams
```

### 03. 從 nba 資料庫的球員資料表 players 中選擇 firstName、lastName 兩個變數，並依序取別名為 first_name、last_name，並只顯示前五列資料。
* 練習 AS 的使用 - 修改取出資料的欄位名稱 

```python
select_variables_as_from_players =\
"""
-- SQL 查詢語法起點

SELECT firstName AS first_name,
       lastName AS last_name
  FROM players
 LIMIT 5;

-- SQL 查詢語法終點
"""

variables_as_from_players = pd.read_sql(select_variables_as_from_players, conn)
variables_as_from_players
```

### 04. 從 twElection2020 資料庫的 admin_regions 資料表選擇「不重複」的縣市（county）
* 練習 DISTINCT 的使用 : 回傳取出資料的不重複數值

```python
select_distinct_county_from_admin_regions =\
"""
-- SQL 查詢語法起點

SELECT DISTINCT county
  FROM admin_regions;

-- SQL 查詢語法終點
"""

distinct_county_from_admin_regions = pd.read_sql(select_distinct_county_from_admin_regions, conn)
distinct_county_from_admin_regions
```

### 05. 從 nba 資料庫的 teams 資料表選擇「不重複」的分組（divName）
* 練習 DISTINCT 的使用

```python
select_distinct_division_from_teams =\
"""
-- SQL 查詢語法起點

SELECT DISTINCT divName
  FROM teams;

-- SQL 查詢語法終點
"""

distinct_division_from_teams = pd.read_sql(select_distinct_division_from_teams, conn)
distinct_division_from_teams
```
