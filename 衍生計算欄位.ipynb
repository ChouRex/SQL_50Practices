### 06. 從 covid19 資料庫的 daily_report 資料表根據 Confirmed、Recovered、Deaths 欄位以及下列公式衍生計算欄位 Active
* Active = Confirmed − Recovered − Deaths
* 練習透過加減來創造出新的欄位

```python
calculate_active_from_daily_report =\
"""
-- SQL 查詢語法起點

SELECT Confirmed,
       Recovered,
       Deaths,
       Confirmed - Recovered - Deaths AS Active
  FROM daily_report


-- SQL 查詢語法終點
"""

active_from_daily_report = pd.read_sql(calculate_active_from_daily_report, conn)
active_from_daily_report
```

### 07. 從 nba 資料庫的 players 資料表依據 heightMeters、weightKilograms 欄位以及下列公式衍生計算欄位 bmi
* BMI = weight / height^2
* 練習透過乘除來創造出新的欄位

```python
calculate_bmi_from_players =\
"""
-- SQL 查詢語法起點

SELECT heightMeters,
       weightKilograms,
       weightKilograms / (heightMeters * heightMeters) AS bmi
  FROM players;

-- SQL 查詢語法終點
"""

bmi_from_players = pd.read_sql(calculate_bmi_from_players, conn)
bmi_from_players
```

### 08. 從 nba 資料庫的 teams 資料表連接 confName、divName 欄位後並去除重複值
* 練習使用 || || 文字運算符來連結 Text

```python
distinct_concatenate_conference_division_from_teams =\
"""
-- SQL 查詢語法起點

SELECT DISTINCT confName || ', ' || divName AS conf_div
  FROM teams;

-- SQL 查詢語法終點
"""

concatenate_conference_division_from_teams = pd.read_sql(distinct_concatenate_conference_division_from_teams, conn)
concatenate_conference_division_from_teams
```
